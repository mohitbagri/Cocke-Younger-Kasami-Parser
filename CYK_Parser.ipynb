You do NOT need to modify the cell below. The function binarizes a grammar in non-CNF and outputs the set of grammer/lexical rules in CNF along with the associated probabilities for PCKY parsing. 




```
def genGrammar(ruleFile):

    rules = []
    mastr = []
    with open(ruleFile) as f :
        for line in f:
            rule = line.strip().split('\t')
            rules.append([rule[1], rule[2:], float(rule[0])])
            mastr.append([rule[1], rule[2:]])
            
    cnf = []           
    while len(rules) > 0 :
        [lhs, rhs, p] = rules.pop()
        if len(rhs) <= 2: cnf.append([lhs, rhs, p])
        else :
            rest = rhs[:len(rhs)-1]
            first = '-'.join(rest)
            if [first, rest] not in mastr :
                rules.append([first, rest, 1])
                mastr.append([first, rest])
            cnf.append([lhs, [first, rhs[-1]], p])

    mysteryVariable = {}
    for [lhs, rhs, p] in cnf :
        if lhs not in mysteryVariable :
            mysteryVariable[lhs] = p
        else :
            mysteryVariable[lhs] += p

    cleanRules = []
    for r in cnf:
        cleanRules.append((r[0], r[1], r[2]/mysteryVariable[r[0]]))
    
    return cleanRules

```

### 1.4: Implement a probabilistic CYK parser

Given a probabilistic context-free grammar and lexicon and an input string, the algorithm should return the highest probability parse tree for that input string along with its probability. When storing backpointers to the two "best" constituents that make the new symbol, we suggest you using nested lists. Here is an example of how a parse tree might look like:

['S', ['NP', ['PRP$', 'Her'], ['NN', 'creditor']], ['VP', ['VBZ', 'makes'], ['NP', ['DT', 'a'], ['NN', 'barrier']]]]

```
def PCYK(sent, rules_dict):
    #Input parameters: sent: the input sentence as a list of strings
    #                  rules: the set of rules as a list of tuples of the form (left-hand-side, right-hand-side, probability)
    #Output parameters: If the sentece is in the language, return a tuple containing the highest probability parse tree along with its probability
    #                   Otherwise, return None
    
    # create the chart
    T = {}
    B_parse = {}
    # INSERTING EXTRA WORD AT THE START OF THE SENTENCE SO THAT WE CAN INDEX BY 1
    sent.insert(0,"%")

    for i in range(1,len(sent)):
      for A in left_hand_side_list:
        rule_key = A + '*'+ sent[i]
        # FINDING PROBABILITY OF RULE A->G
        prob = rules_dict.get(rule_key,'NA')
        if prob =='NA':
          prob=0
        else:
          # BASE CASE FOR BACK POINTERS 
          B_parse_key = str(i-1) + '*' + str(i) + '*' + A
          B_parse[B_parse_key] = sent[i]
        key = str(i-1) + '*' + str(i) + '*' + A
        T[key] = prob
      for j in range(i-2, -1, -1):
          for k in range(j+1,i):
            for grammer_rule in grammar_rules:
              A= grammer_rule[0]
              B= grammer_rule[1][0]
              C= grammer_rule[1][1]
              key1= str(j) + '*' + str(k) + '*' + B
              key2 = str(k) + '*' + str(i) + '*'+ C

              x= T.get((key1),0)
              y= T.get((key2),0)

              P = grammer_rule[2]

              pr = x *y * P
              new_key = str(j) + '*' + str(i) + '*' + A
              new_prob = T.get(new_key,0)
              if pr > new_prob:
                T[new_key] = pr
                B_parse_key1 = str(j) + '*' + str(k) + '*' + B
                B_parse_key2 = str(k) + '*' + str(i) + '*' + C

                value1=[]
                value1.append(B)
                value1.append(B_parse[B_parse_key1])

                value2=[]
                value2.append(C)
                value2.append(B_parse[B_parse_key2])

                temp_list=[]

                temp_value1=[]
                temp_value2=[]

                # FEW PROCESSING STEPS TO MAKE THE FINAL PARSE TREE IN THE CORRECT FORM SO THAT IT CAN BE PRINTED BY PPRINT 

                if isinstance(value1[1], str):
                  temp_list.append(value1)
                else:
                   temp_value1.append(value1[0])
                   temp_value1.append(value1[1][0])
                   temp_value1.append(value1[1][1])
                   temp_list.append(temp_value1)
                if isinstance(value2[1], str):
                  temp_list.append(value2)
                else:
                   temp_value2.append(value2[0])
                   temp_value2.append(value2[1][0])
                   temp_value2.append(value2[1][1])
                   temp_list.append(temp_value2)
                B_parse[new_key] = temp_list               

    # CHECKING IF KEY OF FORM 0,N,S EXISTS IN T OR NOT (i.e. WHETHER THE SENTENCE IS VALID OR NOT)    
    final_key= '0' + '*' + str(len(sent)-1) + '*' + 'S' 
    final_prob = T.get(final_key,0)

    if final_prob > 0:
      return  (B_parse[final_key], T[final_key])
    else:
      return None

```

```
# HELPER METHOD TO PRINT PARSE TREE 
def printTree(s):
  final_list=[]
  final_list.append('S')
  final_list.extend(s)
  pprint(final_list, indent=7, width=5)
```
